```python
from sklearn.datasets import load_wine

wine = load_wine(as_frame=True)
```

```python
print(wine.DESCR)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    wine.data, wine.target, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC, SVC

pipl = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(random_state=42))
])

pipl.fit(X_train, y_train)

# Predict on the test data
y_pred = pipl.predict(X_test)

# Evaluate the performance
from sklearn.model_selection import cross_val_score

cross_val_score(pipl, X_train, y_train).mean()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform, uniform

param_distributions = {
    "svm__gamma": loguniform(0.001, 0.1),
    "svm__C": uniform(1, 10)
}

rnd = RandomizedSearchCV(pipl, param_distributions=param_distributions , cv=5, n_iter=100, random_state=42)
rnd.fit(X_train, y_train)

rnd.best_score_
```

```python
rnd.score(X_test, y_test)
```
