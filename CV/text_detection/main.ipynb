```python
from easyocr import Reader


cap = cv2.VideoCapture(0)
reader = Reader(['en', 'uk'])

while True:
    ret, frame = cap.read()
    cv2.imshow('pacan', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('r'):
        results = reader.readtext(frame)
        detected = frame
        
        for cords, text, prob in results:
            print(f"{text}, - {prob}")
            print((cords[0][0], cords[0][1]), (cords[2][0], cords[2][1]), (255, 0, 0), 5)
            detected = cv2.rectangle(detected, 
                                     (int(cords[0][0]), int(cords[0][1])), 
                                     (int(cords[2][0]), int(cords[2][1])), 
                                     (255, 0, 0), 5)
            
        cv2.imshow('recog', detected)
    
    elif cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
